```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from luwiji.pandas import illustration
```

# Statistical & Summary

```python
df = pd.read_csv("data/cereal.csv")
df.head()
```

```python
illustration.nomenklatur
```

## Missing value

```python
plt.figure(figsize=(6, 6))
sns.heatmap(df.isna(), cbar=False, cmap="Blues")
```

## Corralation Heatmap

```python
plt.figure(figsize=(16,10))
sns.heatmap(df.corr(), 
            vmin=-1, 
            vmax=1, 
            annot=True, 
            cmap="bwr");

# Semakin merah semakin berkorelasi
```

## Categorical Data

```python
sns.catplot(x="mfr", y="rating", data=df) # sebaran data
plt.xlabel("Manufaktur")
```

### Counting Visualization

```python
sns.countplot(x="mfr", 
              data=df, 
              hue="shelf"
             );
plt.xlabel("Manufaktur")
plt.ylabel("Banyak Produk")

# Manufaktur yang banyak memproduksi adalah "Kellogs" 
# dan lebih banyak menyimpan produknya di rak ke-3
```
