```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 50)
y1 = np.sin(X)
y2 = np.cos(X)
```

```python
plt.figure(figsize=(4,4))
plt.title("SINUS & COSINUS")
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.plot(X, y1, 'g', linewidth=0.5, label="Sinus")
plt.plot(X, y2, 'r', linewidth=0.3, label="Cosinus")
plt.legend(loc="lower left");
```

## Scatter plot

```python
df = pd.read_csv("data/circle_data.csv", sep=";")
df.head()
```

```python
X, y, rad = df.x, df.y, df.radius
```

```python
# c = color
# cmap = color map
plt.scatter(X, y, c=rad, s=rad*100, cmap="seismic")
plt.axis("equal");
```

## Histogram

```python
data = np.random.randn(100)

plt.hist(data, bins=10);
```

```python
plt.hist(data, bins=15, density=1, cumulative=True);
```

## Subplot

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,7))
ax[0,0].scatter(X, y, c=rad, s=rad*50, cmap="seismic")
ax[1,1].hist(data, bins=15, density=1, cumulative=True)
ax[1,0].hist(data, bins=5);
```
