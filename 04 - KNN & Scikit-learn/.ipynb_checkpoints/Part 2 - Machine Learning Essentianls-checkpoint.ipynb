## Menentukan Feature/Input (X) dan Target/Label (y)

```python
# Encoding Categorical Data -- OneHot Encoded and -- and drop uncorrelation
X = pd.get_dummies(df, columns=["Sex", "Pclass", "Embarked"]).drop(columns=["Survived", "SibSp", "Parch"])
y = df["Survived"]  # Selamat atau tidak adalah sebagai target prediksi
X
```

## Fitting

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3, 
                             metric="manhattan",
                             weights="distance"
        )
model.fit(X, y)
```

### Cek keakuratan Prediksi

```python
score = model.score(X, y) * 100
print(f"Tingkat akurasi : {score} %")
```

### Dikarenakan datanya bersifat Classification - Gunakan 'Stratify' saat splitting data set

```python
# Encoding Categorical Data -- OneHot Encoded and -- and drop uncorrelation
X = pd.get_dummies(df, columns=["Sex", "Pclass", "Embarked"]).drop(columns=["Survived", "SibSp", "Parch"])
y = df["Survived"]  # Selamat atau tidak adalah sebagai target prediksi

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y ,random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

test_model = KNeighborsClassifier(n_neighbors=3, weights="distance")
test_model.fit(X_train, y_train)

model = KNeighborsClassifier(n_neighbors=3, weights="distance")
model.fit(X, y)
```

```python
test_score_model = test_model.score(X_test, y_test) * 100
score_model = model.score(X, y) * 100

if ( score_model - test_score_model ) > 10:
    print(f'model: {score_model} \nmodel_test: {test_score_model} \n\nOverfitting Model')
else:
     print(f'model: {score_model} \nmodel_test: {test_score_model} \n\nNot Overfitting Model')
```
