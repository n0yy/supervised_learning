# ENCODING 
    merubah data kategorikal menjadi numerikal

 - label encoding
	digunakan saat datanya bersifat orinalitas
	ex: gender (male/female) jadi 0/1
	    tingkat kepuasan (sangat buruk/buruk/baik/sangat baik) jadi 0/1/2/3

 - one hot encoding	
	digunakan kapanpun, karena lebih umum
	ex:
		|rak 	|

		|bawah	|
		|tengah	|	
		|atas	|
		

		jadi

		isBawah | isTengah	| isAtas
		   1		0	    0
		   0		1	    0
		   0		0	    1

## One Hot Encoding / Dummy Variable

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/cereal_imute.csv")
df.tail()
```

```python
pd.get_dummies(df, columns=["shelf"], prefix="rak", prefix_sep="__")
```

## Binning Numerical
Mengubah menjadi numerik menjadi kategorik

```python
df["rating_category"] = pd.cut(df.rating, 
                               bins=[0, 25, 50, 75, df.rating.max()], 
                               labels=["buruk", "cukup", "baik", "sangat baik"], )
df
```

```python
# df.rating.hist(bins=[0,25,50,df.rating.max()])
df.rating.hist(bins=4)
```
